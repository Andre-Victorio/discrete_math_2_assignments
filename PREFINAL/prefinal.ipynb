## Directed Graph

This is a directed graph. The reason this is called a directed graph is because the edges are directed, meaning they have a direction.

```python
import matplotlib.pyplot as mx
import networkx as nt
seed = 65
G = nt.DiGraph()
G.add_nodes_from([0,1,2,4,7])
edgelist=[(0,2),(2,4),(4,1),(1,7),(7,0)]
G.add_edges_from(edgelist)
pos = nt.circular_layout(G)
nt.draw(G,pos=pos,with_labels=True)
mx.show()
```

## Simple Graphs

This is a Simple graph, meaning there are no loops or parallel edges in this graph. 

```python
import matplotlib.pyplot as mx
import networkx as nt
seed = 65
G = nt.Graph()
G.add_nodes_from([0,1,2,4,7])
edgelist = [(2,4),(4,1),(1,7),(0,7)]
G.add_edges_from(edgelist)
pos = nt.circular_layout(G)
nt.draw(G,pos=pos,with_labels=True)
mx.show()
```

## Complete Graphs

```python
import matplotlib.pyplot as mpl
import networkx as nt

G = nt.Graph()
G.add_nodes_from([0,1,4,5,7,9])
edges = [(0,5),(5,1),(4,7),(9,1),(4,5)]
G.add_edges_from(edges)
pos = nt.planar_layout(G)
nt.draw(G,pos=pos,with_labels=True)
mpl.show()
```

## Bipartite Graph

```python

import matplotlib.pyplot as mpl
import networkx as nt

G = nt.Graph()
G.add_nodes_from([0,1,4], bipartite=0)
G.add_nodes_from([2,3,6], bipartite=1)
G.add_edges_from([(0,2),(4,3),(1,6),(0,3),(0,6),(1,3),(1,2),(4,2),(4,6)])

top = nt.bipartite.sets(G)[0]
pos = nt.bipartite_layout(G,top)
nt.draw(G,pos=pos,with_labels=True)
mpl.show()
```

## Euler's Circuit
An Euler's Circuit or Eulerian Circuit is a type of graph connectivity theorem that graph will contain an Eulerian Circuit if a graph is connected and every vertex has a positive even number of degrees or number of edges connected to it.

```python
import matplotlib.pyplot as mpl
import networkx as nt

G = nt.Graph()
G.add_nodes_from([0,1,2,3,4,5])
G.add_edges_from([(0,1),(0,3),(0,2),(0,5),(1,2),(1,3),(1,5),(2,3),(3,4),(2,4)])
pos = nt.shell_layout(G)
nt.draw(G,pos=pos,with_labels=True)
mpl.show()
```

## Euler's Trail

Euler's Trail is a sequence of adjacent edges and vertices that that starts from a point $v$ to an end point $w$. Unlike Euler's Circuit, the start and end points will have odd degrees, while the rest of the vertices will have even degrees.

```python

import matplotlib.pyplot as mpl
import networkx as nt

G = nt.Graph()
G.add_nodes_from([0,1,2,3,4])
G.add_edges_from([(0,1),(0,2),(0,4),(1,2),(2,4),(2,3),(3,4)])
pos = nt.planar_layout(G)
nt.draw(G,pos=pos,with_labels=True)
mpl.show()
```

## Hamiltonian Circuit
Hamiltonian Circuits are circuits that pass through each vertex in Graph $G$ once and only once. A complete Graph with $N$ is $(N-1)!$

```python


import matplotlib.pyplot as mpl
import networkx as nt

G = nt.Graph()
G.add_nodes_from([0,1,2,3,4])
G.add_edges_from([(0,1),(0,2),(0,4),(1,3),(1,2),(1,4),(2,4),(2,3),(3,4),(0,3)])
pos = nt.shell_layout(G)
nt.draw(G,pos=pos,with_labels=True)
mpl.show()
```
